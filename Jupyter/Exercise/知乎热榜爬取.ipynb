# 知乎热榜爬取（通过今日热榜源）

```python
import requests
import re
url='https://tophub.today/n/mproPpoq6O'
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.116 Safari/537.36'
}
s=requests.get(url,headers=headers)
#print(s.text)
```

```python
example = '<a href=".*?" target="_blank" rel="nofollow" itemid=".*?">(.*?)</a>'
pattern = re.compile(example, re.S)
titles = re.findall(pattern, s.text)
print(titles)
```

```python
import numpy as py
import pandas as pd
ser1=pd.Series(titles,index=[i for i in range(len(titles))])
df1=pd.DataFrame(ser1,columns=['titles'])
df1
```

```python
df1.to_csv('知乎热榜爬取')
```
