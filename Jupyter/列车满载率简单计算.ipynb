```python
#以下为计算一段时间内经过一个站点的列车班次数
```

```python
def cut(data):
    position = data.find(' ')
    data2 = data[:position]+' 00:00:00'
    result = time.mktime(time.strptime(data, "%Y-%m-%d %H:%M:%S")) - \
        time.mktime(time.strptime(data2, "%Y-%m-%d %H:%M:%S"))
    return result
```

```python
filename = r"D:\Data\AFC_客流数据\20150407_transfer.csv"
trip = pd.read_csv(filename)
```

```python
data = trip[(trip['outLine'] == 5)&(trip['inLine']==5)]
data
```

```python
data["transfered_inTime"] = data["inTime"].apply(
    lambda x: cut(x))
data["transfered_outTime"] = data["outTime"].apply(
    lambda x: cut(x))
```

```python
data['inTime'] = pd.to_datetime(data['inTime'])
data['outTime'] = pd.to_datetime(data['outTime'])
data['dayName'] = data['inTime'].dt.dayofweek
data['dayName'] = data['outTime'].dt.dayofweek
```

```python
replace = {
    "莘庄":111,
"春申路":502,
"银都路":503,
"颛桥":505,
"北桥":507,
"剑川路":508,
"东川路":509,
"金平路":510,
"华宁路":511,
"文井路":512,
"闵行开发区":513}
```

```python
data=data.replace(replace)  
```

```python
data_502 = data[(data['outStation'] == 502)]
data_502_down = data_502.sort_values(by=['transfered_outTime'], ascending=True)
```

```python
data_502_down
```

```python
a = [23173.0]
n = len(data_502_down)
##前一个乘客的出站时间与当前乘客相差不超过K=【60、、、、】
##添加最早出站班次
for i in range(n-1):
    if data_502_down.iloc[i+1]['transfered_outTime'] - data_502_down.iloc[i]['transfered_outTime'] >60:
        a.append(data_502_down.iloc[i+1]['transfered_outTime'])
```

```python
def transform_time(seconds):
# 得到总秒数
    #seconds = float(seconds)*60
    m, s = divmod(seconds, 60)
    h, m = divmod(m, 60)
    time = "%d:%02d:%02d" % (h, m, s)
    return time
```

```python
b = []
for i in range(0,len(a)):
    b.append(transform_time(a[i]))
```
