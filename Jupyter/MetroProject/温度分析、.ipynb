```python
data_rain['intime_normal'] = data_rain['inTime'].apply(lambda x: time.mktime(time.strptime(x,"%Y-%m-%d %H:%M:%S")))
data_rain['timecs'] = data_rain['outtime_normal']-data_rain['intime_normal']
```

```python
rain = np.array([24,32,10,2,0,1])
sun = np.array([21,25,11,1,0,0])
plt.figure(figsize = (10,8),dpi = 1000)
plt.xlabel('Seconds')
plt.ylabel('Number of traveling trips')
plt.tick_params(labelsize=10)
plt.grid(linestyle=':', axis='y')
x = np.arange(6)
a = plt.bar(x - 0.2, rain, 0.4,  label='April,9,2015(10°c-13°c)', align='center')
b = plt.bar(x + 0.2, sun, 0.4,  label='April,23,2015(13°c-25°c)', align='center')
# 设置标签
for i in a + b:
    h = i.get_height()
    plt.text(i.get_x() + i.get_width() / 2, h, '%d' % int(h), ha='center', va='bottom')
plt.xticks(x, ['3000', '3300', '3600', '3900', '4200', '4500'])

plt.legend()
plt.show()
```
