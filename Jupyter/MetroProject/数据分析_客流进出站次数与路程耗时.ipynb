```python
def StrtoTimestamp(time_str):
    time_stamp = time.mktime(time.strptime(time_str, "%Y-%m-%d %H:%M:%S"))
    return time_stamp
```

```python
# pd:pandas 数据
# colName 列名
# timeInterval 时间间隔
def regionCount(pd, colName, timeInterval,startUnix,maxUnix):
    result = []
    while startUnix:
        endUnix = startUnix + timeInterval
        tempBool = (startUnix <= pd[colName]) &  (pd[colName] < endUnix)
        result.append(pd[tempBool])
        startUnix = startUnix + timeInterval
        if startUnix > maxUnix:
            break
    return result
```

```python
import pandas as pd
import glob
 
def hebing(csv_list, outputfile):
    for inputfile in csv_list:
        f = open(inputfile)
        data = pd.read_csv(f)
        data.to_csv(outputfile, mode='a', index=False)
    print('完成合并')
```

```python
def quchong(file):
    df = pd.read_csv(file, header=None)
    datalist = df.drop_duplicates()
    datalist.to_csv('result_new.csv', index=False, header=False)
    print('完成去重')
```
