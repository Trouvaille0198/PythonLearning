```python
temp=atri_data_0407['Temp. (ÂºC)'].tolist()
cond=atri_data_0407['Conditions'].tolist()
_temp=[]
_cond=[]
```

```python
temp
```

```python
for i in range(len(line5_traveler_data.index)):
    count=7200
    if(line5_traveler_data.iloc[i,10]<7200 or line5_traveler_data.iloc[i,10]>82800):
        _temp.append(temp[0])
        _cond.append(cond[0])
    else:
        for j in range(7,0,-1):
            if(line5_traveler_data.iloc[i,10]>=count and line5_traveler_data.iloc[i,10]<=count+10800):
                _temp.append(temp[j])
                _cond.append(cond[j])
                break
            count=count+10800
```

```python
line5_traveler_data['temp']=_temp
line5_traveler_data['conditions']=_cond
```
